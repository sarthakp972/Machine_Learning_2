#Binning :- neumercal data convert into catagorize data.
binnig refers to the process of grouping a set of continous or numerical data points into a smaller number of discrete "bins" or intervals . binning is oiften used as adata preprocessing step to transform continous fetures into categorical ones

for column horsepower-power of the Engine

```python
df1['horsepower']=df1['horsepower'].astype(float)
df1['horsepower'].dtype
```

```python
#hist -Histogram provide a visual summary of the underlying frequency dis
plt.figure(figsize=(4,2))
plt.hist(df1['horsepower'])
plt.xlabel('Horsepower values')
plt.ylabel('frequency')
```

```python
np.linspace(1,10,4)#Ex:divide into 4 parts
```

```python
bins = np.linspace(min(df1['horsepower']), max(df1['horsepower']),4)#4= 4 parts ,
bins
```

```python
group_names=['low', 'medium','high']
```

```python
df1["horsepower-binned"] = pd.cut(df1['horsepower'],bins,labels=group_names,include_lowest=True)
```

```python
df1[['horsepower', 'horsepower-binned']].head(20)
```

```python
df1['horsepower-binned'].value_counts()
```

```python
plt.figure(figsize=(4,2))
plt.bar(group_names,df1['horsepower-binned'].value_counts())
plt.xlabel('horsepower')
plt.ylabel('frequecy')
```
